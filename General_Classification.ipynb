```python
def target_corr(data, target, threshold=0.1):
    corr_matrix=data.corr(method='spearman', numeric_only=True)
    target_correlation=corr_matrix[target].abs()>threshold
    var_exp_target=target_correlation[target_correlation].index.tolist()
    var_exp_target.remove(target)

    result=pd.DataFrame({
    'variable':var_exp_target,
    'correlation':corr_matrix.loc[var_exp_target, target]
  })
    result.reset_index(drop=True, inplace=True)
    return result
```

```python
target=target_corr(data_logreg, 'default')
target
```

```python
def intercorr(data, target, threshold=0.7):
    corr_matrix=data.corr(method='spearman', numeric_only=True)
    high_corr_var=(corr_matrix.abs()>threshold) & (corr_matrix!=1)
    independent_var=np.where(high_corr_var)
    correlated_pairs=set()
    for var1, var2 in zip(independent_var[0], independent_var[1]):
        if var1<var2:
            correlated_pairs.add((corr_matrix.index[var1], corr_matrix.columns[var2]))
        else:
            correlated_pairs.add((corr_matrix.index[var2], corr_matrix.columns[var1]))
    result=pd.DataFrame(list(correlated_pairs), columns=['varieble1', 'variable2'])
    return result

```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def vif_result(data, variables):
    vif_data=data[variables]
    vif_values=[variance_inflation_factor(vif_data.values, i) for i in range(vif_data.shape[1])]
    vif=pd.DataFrame({'VIF':vif_values,
                    'Feature':variables})
    return vif
```

```python
for i in data_logreg.columns[:-1]:
    if data_logreg[i].dtypes!=object:
        q1=data[i].quantile(0.25)
        q2=data[i].quantile(0.5)
        q3=data[i].quantile(0.75)

        ranges=[-np.inf, q1, q2, q3, np.inf]
        data_logreg['category_'+i]=pd.cut(data_logreg[i], bins=ranges, duplicates='drop')
        grouped=data_logreg.groupby(['category_'+i, 'default'])['default'].count().unstack().reset_index()
        grouped['positive_prob']=grouped[0]/grouped[0].sum()
        grouped['negative_prob']=grouped[1]/grouped[1].sum()
        grouped[i+'_woe']=np.log(grouped['positive_prob']/grouped['negative_prob'])
        data_logreg=pd.merge(data_logreg, grouped[['category_'+i, i+'_woe']], on='category_'+i, how='left')
```

```python
for var in data_logreg.columns:
    if data_logreg[var].dtypes==object:
        grouped=data_logreg.groupby([var, 'default'])['default'].count().unstack().reset_index()
        grouped['positive_prob']=grouped[0]/grouped[0].sum()
        grouped['negative_prob']=grouped[0]/grouped[0].sum()
        grouped[var+'_woe']=np.log(grouped['positive_prob']/grouped['negative_prob'])
        data_logreg=data_logreg.merge(grouped[[var, var+'_woe']], on=var, how='left')
```

```python
from sklearn.linear_model  import LogisticRegression
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
from sklearn.metrics import roc_auc_score
from sklearn import metrics
```

```python
def train_and_evaluate1(model_name, model, x_train, y_train, x_test, y_test):
    model.fit(x_train, y_train)
    y_pred=model.predict(x_test)
    y_prob_test=model.predict_proba(x_test)[:,1]
    roc_test=roc_auc_score(y_test, y_prob_test)
    gini_test=(2*roc_test)-1
    
    print(f'Model performance for {model_name}')
    print(f'Gini is {gini_test*100}')
    return gini_test
```

```python
models=[]

LR_def=LogisticRegression()
RF_def=RandomForestClassifier()
XGB_def=XGBClassifier()
LGBM_def=LGBMClassifier()
CB_def=CatBoostClassifier()
CB_cat_def=CatBoostClassifier(cat_features=['SEX', 'EDUCATION', 'MARRIAGE'])
SVC_def=SVC(probability=True)

models.extend([('LogisticRegression_default', LR_def),
              ('RandomForest_default', RF_def),
              ('XGBoost_default', XGB_def),
              ('LightGBM_default', LGBM_def),
              ('CatBoost_default', CB_def),
              ('CatBoost_categoric_default', CB_cat_def),
              ('SVClassifier_default', SVC_def)])
```

```python
gini_df=pd.DataFrame(columns=['model', 'gini'])

for model_name, model in models:
    if model_name=='LogisticRegression_default':
        gini=train_and_evaluate1(model_name, model, x_train_log, y_train_log, x_test_log, y_test_log)
    elif model_name=='CatBoost_categoric_default':
        gini=train_and_evaluate1(model_name, model, x_train_cat, y_train_cat, x_test_cat, y_test_cat)
    else:
        gini=train_and_evaluate1(model_name, model, x_train_dummy, y_train_dummy, x_test_dummy, y_test_dummy)
    if gini is not None:
        gini_df=pd.concat([gini_df, pd.DataFrame({'model':[model_name], 'gini':[gini]})], ignore_index=True)
        
gini_df_sorted=gini_df.sort_values(by='gini', ascending=False)
gini_df_sorted
```

```python
import optuna
from sklearn.model_selection import cross_val_score
```

```python
def best_params_for_model(trial):
    param = {
    'n_estimators' : trial.suggest_int('n_estimators', 100, 1000, step=100),
    'max_depth' : trial.suggest_int('max_depth', 3, 20),
    'min_samples_split' : trial.suggest_int('min_samples_split', 2, 20),
    'min_samples_leaf' : trial.suggest_int('min_samples_leaf', 1, 10),
    'max_features' : trial.suggest_categorical('max_features', [ 'sqrt', 'log2'])
    }

    rf_clf = RandomForestClassifier(**param)

    auc = cross_val_score(rf_clf, x_train_dummy, y_train_dummy, cv=3, scoring='roc_auc', n_jobs=-1).mean()
    return auc

study = optuna.create_study(direction='maximize')
study.optimize(best_params_for_model, n_trials=1)

print('Best trial:')
rf_params = study.best_params
print('  Value: {:.3f}'.format(study.best_value))
print('  Params: ', rf_params)
RF_optuna = RandomForestClassifier(**rf_params)
```

```python
def best_params_for_model(trial):
    param = {
        'n_estimators': trial.suggest_int('n_estimators', 10, 1000),
        'learning_rate': trial.suggest_loguniform('learning_rate', 0.01, 1.0),
        'max_depth': trial.suggest_int('max_depth', 3, 10),
        'subsample': trial.suggest_uniform('subsample', 0.5,1),
        'colsample_bytree': trial.suggest_uniform('colsample_bytree', 0.5,1),
        'gamma': trial.suggest_int('gamma', 0,5)

}
    xgb_clf = XGBClassifier(**param)

    auc = cross_val_score(xgb_clf, x_train_dummy, y_train_dummy, cv=3, scoring='roc_auc', n_jobs=-1).mean()
    return auc



study = optuna.create_study(direction='maximize')
study.optimize(best_params_for_model, n_trials=10)

print('Best trial:')
xgb_params = study.best_params
print('  Value: {:.3f}'.format(study.best_value))
print('  Params: ', xgb_params)

xgb_optuna = XGBClassifier(**xgb_params)
```
